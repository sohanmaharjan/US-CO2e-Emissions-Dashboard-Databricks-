# Emissions in US states Query

### Location Data

```python
%sql
SELECT latitude,longitude,`GHG emissions mtons CO2e` as Emissions
FROM emissions_data
```

### Emissions per person

```python
%sql
SELECT
  county_state_name,
  population,
  CAST(REPLACE(`GHG emissions mtons CO2e`, ',','') AS DOUBLE) / 
  NULLIF(CAST(REPLACE(population, ',','') AS DOUBLE),0)
  AS emissions_per_person
FROM
  emissions_data
ORDER BY
  emissions_per_person DESC
```

### Total Emissions per State

```python
%sql
SELECT
  state_abbr,
 SUM(CAST(REPLACE(`GHG emissions mtons CO2e`, ',','') AS DOUBLE))
  AS TOTAL_EMISSIONS
FROM
  emissions_data
GROUP BY
  state_abbr
ORDER BY
  TOTAL_EMISSIONS DESC
LIMIT 10
```

### County Shaming

```python
%sql
SELECT
  county_state_name,
  population,
  CAST(REPLACE(`GHG emissions mtons CO2e`, ',','') AS DOUBLE)
  AS Total_Emissions
FROM
  emissions_data
ORDER BY
  Total_Emissions DESC
LIMIT 10
```
